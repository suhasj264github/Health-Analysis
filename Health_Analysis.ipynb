```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

df = pd.read_csv('medical_examination.csv')
df
```

```python
df.isnull().sum() 
```

```python
df['height'] = df['height']/100
df['BMI'] = df['weight']/df['height']**2
ok = [] 
for w in df['BMI']:
    if w>25:
         ok.append(1)
    else:
        
        ok.append(0)
df['Overweight'] = ok
df
```

```python
df[['cholesterol', 'gluc']] 
```

```python
df[['cholesterol', 'gluc']] = np.where(df[['cholesterol', 'gluc']]==1, 0, df[['cholesterol', 'gluc']])
df[['cholesterol', 'gluc']] = np.where(df[['cholesterol', 'gluc']]>1, 1, df[['cholesterol', 'gluc']])
df[['cholesterol', 'gluc']]
```

```python
df
```

```python
df2 = df.melt(id_vars= ['cardio'], value_vars=['active', 'alco', 'cholesterol', 'gluc', 'Overweight', 'smoke'])
df2
```

```python
"df2['value_counts'] = df2['value'].value_counts()\na = df2['value_counts']\ndf2.dropna()"
sns.catplot(data = df2, x= 'variable', kind='count', hue='value', col='cardio')
```

```python
df.drop('BMI', axis=1, inplace=True)

df = df[df['ap_lo'] <= df['ap_hi']]
df = df[df['height'] >= df['height'].quantile(0.025)]
df = df[df['height'] <= df['height'].quantile(0.975)]
df = df[df['weight'] >= df['weight'].quantile(0.025)]
df = df[df['weight'] <= df['weight'].quantile(0.975)]

df_corr = df.corr()

df_corr
```

```python
df
```

```python
df['age'] = df['age']/365
df
```

```python
plt.style.use('dark_background')
plt.style.use('dark_background')
g = sns.catplot(x="weight", y="age", kind='bar', data=df)
g.figure.set_size_inches(90,20)
plt.xticks(rotation='vertical', size=20)
plt.yticks(size=20)
g.fig.suptitle("Categorial plot for age and weight", size=50)
g.set_axis_labels("Weight", 'Age', size=50)
plt.show()
```

```python
g = sns.catplot(x="weight", y="ap_hi", kind='bar', data=df)
g.figure.set_size_inches(90,20)
plt.xticks(rotation='vertical', size=20)
plt.yticks(size=20)
g.fig.suptitle("Categorial plot for Age and Systolic blood pressure", size=50)
g.set_axis_labels("Weight", 'Age', size=50)
plt.show()
```

```python
g = sns.catplot(x="weight", y="ap_lo", kind='bar', data=df)
g.figure.set_size_inches(90,20)
plt.xticks(rotation='vertical', size=20)
plt.yticks(size=20)
g.fig.suptitle("Categorial plot for Age and Diastolic blood pressure", size=50)
g.set_axis_labels("Weight", 'Age', size=50)
plt.show()
```

```python
mask = np.zeros_like(df_corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style('dark'):
    f, ax = plt.subplots(figsize=(16, 12))
    g = sns.heatmap(df_corr, mask=mask, annot = True, annot_kws={'size': 8}, vmax= .24, fmt='0.1f')
```
